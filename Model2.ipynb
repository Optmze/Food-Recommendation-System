```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
data = pd.read_csv("foodsfinal2.csv")

# Select relevant columns for similarity calculation
selected_columns = ["calories", "fat_total_g", "protein_g", "carbohydrates_total_g", "fiber_g", "sugar_g"]
nutritional_data = data[selected_columns]

# Compute the cosine similarity matrix based on nutritional data
cosine_sim = cosine_similarity(nutritional_data, nutritional_data)

# Function to get recipe recommendations based on nutritional similarity
def get_nutritional_recommendations(recipe_index, cosine_similarity=cosine_sim, dataset=data, top_n=3):
    # Get the similarity scores
    similarity_scores = list(enumerate(cosine_sim[recipe_index]))
    
    # Sort recipes based on similarity scores
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    # Get top N similar recipes (excluding the queried recipe itself)
    top_similar_recipes = [dataset.iloc[score[0]]["RecipeName"] for score in similarity_scores if score[0] != recipe_index][:top_n]
    
    return top_similar_recipes

# Example usage
recipe_index = 36  # Index of the recipe you want to find similar recipes for
recommendations = get_nutritional_recommendations(recipe_index)

print(f"Recipes similar to '{data.iloc[recipe_index]['RecipeName']}':")
for recommendation in recommendations:
    print(recommendation)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load the dataset
data = pd.read_csv("foodsfinal2.csv")

# Select relevant columns for KNN
selected_columns = ["calories", "fat_total_g", "protein_g", "carbohydrates_total_g", "fiber_g", "sugar_g"]
X = data[selected_columns]

# Fit KNN model
k = 5  # Number of neighbors to consider
knn = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(X)

# Function to recommend similar recipes
def recommend_similar_recipes(selected_recipe):
    # Find nutritional values for the selected recipe
    data["RecipeName"]=data["RecipeName"].tolist()
    selected_recipe_nutrition = data[data["RecipeName"] == selected_recipe][selected_columns]

    if len(selected_recipe_nutrition) == 0:
        print(f"No nutritional information found for '{selected_recipe}'.")
        return []

    # Find k nearest neighbors
    distances, indices = knn.kneighbors(selected_recipe_nutrition)

    # Get indices of similar recipes (excluding the selected recipe)
    similar_indices = indices[0]

    # Get names of similar recipes
    similar_recipes = data.iloc[similar_indices]["RecipeName"].tolist()

    # Remove the selected recipe from the list
    similar_recipes.remove(selected_recipe)

    return similar_recipes

# Example usage
selected_recipe = "Masala Karela Recipe"
recommended_recipes = recommend_similar_recipes(selected_recipe)
print("Recommended recipes:", recommended_recipes)
```
